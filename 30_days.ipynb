```python
#12 days of python 


with open(r"C:\Users\Paramveer Singh\OneDrive\Documents\Portfolio\rotations.txt") as input_rotations:
    rotations = [line.strip() for line in input_rotations]
    pointer = 50
    result = []
    count = 0 


    for r in rotations:
        if r[0] == 'R':
            distance = int(r[1:])
            pointer = (pointer + distance) % 100
            if pointer >= distance:
                count += 1
            result.append(pointer)
            
        elif r[0] == 'L':
            distance = int(r[1:])
            pointer = (pointer - distance) % 100
            if pointer == 0:
                #print ("count", count)
                #print ("distance", distance)
                count += 1
            
            result.append(pointer)

        else:
            print("Invalid rotation command:", r)
        
        print (count)
    
```
